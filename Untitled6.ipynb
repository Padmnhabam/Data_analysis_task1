## Load data

### Subtask:
Load the data from the "/content/B2C_sales_data.csv" file into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the CSV file into a DataFrame, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv('/content/B2C_sales_data.csv')
display(df.head())
```

## Handle missing values

### Subtask:
Identify and handle missing values in the DataFrame. This may involve imputation, deletion, or other strategies depending on the nature and extent of the missing data.


**Reasoning**:
Calculate the percentage of missing values for each column, identify columns with a high percentage of missing values, drop those columns, and then handle the remaining missing values using appropriate strategies. Finally, verify that there are no missing values left.



```python
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().sum() / len(df) * 100

# Identify columns with a high percentage of missing values (e.g., above 70%)
high_missing_cols = missing_percentage[missing_percentage > 70].index.tolist()

# Drop the columns identified in the previous step
df_cleaned = df.drop(columns=high_missing_cols)

# Impute remaining missing values
for column in df_cleaned.columns:
    if df_cleaned[column].isnull().sum() > 0:
        if df_cleaned[column].dtype in ['int64', 'float64']:
            # Impute numerical columns with the median
            df_cleaned[column].fillna(df_cleaned[column].median(), inplace=True)
        elif df_cleaned[column].dtype == 'object':
            # Impute categorical columns with the mode
            df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)

# Verify that there are no remaining missing values
missing_after_cleaning = df_cleaned.isnull().sum().sum()

print("Percentage of missing values before dropping columns:")
print(missing_percentage)
print("\nColumns dropped due to high missing percentage:")
print(high_missing_cols)
print("\nTotal missing values after cleaning:")
print(missing_after_cleaning)

display(df_cleaned.head())
```

## Handle duplicate data

### Subtask:
Identify and remove duplicate rows from the DataFrame.


**Reasoning**:
Identify and remove duplicate rows from the DataFrame and verify the result.



```python
# Identify and remove duplicate rows
rows_before_dropping = df_cleaned.shape[0]
df_cleaned = df_cleaned.drop_duplicates()
rows_after_dropping = df_cleaned.shape[0]

# Verify that there are no duplicate rows remaining
print(f"Number of rows before dropping duplicates: {rows_before_dropping}")
print(f"Number of rows after dropping duplicates: {rows_after_dropping}")
print(f"Number of duplicate rows dropped: {rows_before_dropping - rows_after_dropping}")
```

## Handle outliers

### Subtask:
Detect and address outliers in relevant numerical columns. This might involve removal, transformation, or Winsorizing, depending on the data and the analysis goals.


**Reasoning**:
Identify numerical columns, calculate IQR, define outlier bounds, identify outliers, and Winsorize outliers for each numerical column in df_cleaned. Then print the number of outliers detected and handled.



```python
import numpy as np

# Identify numerical columns
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns

# Handle outliers for each numerical column using Winsorizing
for col in numerical_cols:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
    num_outliers = len(outliers)

    # Winsorize outliers
    df_cleaned[col] = np.where(df_cleaned[col] < lower_bound, lower_bound, df_cleaned[col])
    df_cleaned[col] = np.where(df_cleaned[col] > upper_bound, upper_bound, df_cleaned[col])

    print(f"Column: {col}")
    print(f"Number of outliers detected and handled: {num_outliers}")
    print("-" * 30)

display(df_cleaned.head())
```

## Check data types

### Subtask:
Ensure that the data types of each column are appropriate for analysis. Convert columns to the correct types (e.g., numerical, categorical, datetime).


**Reasoning**:
Display the current data types to identify columns that need conversion.



```python
df_cleaned.info()
```

**Reasoning**:
Convert identified date columns to datetime objects and categorical columns to category type.



```python
# Convert date columns to datetime
date_cols = ['Invoice Date', 'Shipment Date', 'Order Date']
for col in date_cols:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col])

# Convert categorical columns to 'category' dtype
categorical_cols = [
    'Seller Gstin', 'Transaction Type', 'Item Description', 'Asin', 'Sku',
    'Product Tax Code', 'Bill From City', 'Bill From State', 'Bill From Country',
    'Ship From City', 'Ship From State', 'Ship From Country', 'Ship To City',
    'Ship To State', 'Ship To Country', 'Warehouse Id', 'Fulfillment Channel',
    'Payment Method Code'
]
for col in categorical_cols:
    df_cleaned[col] = df_cleaned[col].astype('category')

# Verify the data types after conversion
df_cleaned.info()
```

## Feature engineering (optional but recommended)

### Subtask:
Create new features from existing ones that could be more informative for analysis (e.g., extracting month or year from a date column, creating interaction terms).


**Reasoning**:
Extract year, month, day, and day of the week from 'Invoice Date', calculate shipping time, and display the head of the DataFrame to show the new features.



```python
df_cleaned['Invoice_Year'] = df_cleaned['Invoice Date'].dt.year
df_cleaned['Invoice_Month'] = df_cleaned['Invoice Date'].dt.month
df_cleaned['Invoice_Day'] = df_cleaned['Invoice Date'].dt.day
df_cleaned['Invoice_Day_of_Week'] = df_cleaned['Invoice Date'].dt.dayofweek

df_cleaned['Shipping_Time_Days'] = (df_cleaned['Shipment Date'] - df_cleaned['Order Date']).dt.days

display(df_cleaned.head())
```

## Normalize or scale numerical features (optional)

### Subtask:
Normalize or scale numerical features in the `df_cleaned` DataFrame.


**Reasoning**:
Identify the numerical columns, import StandardScaler, instantiate it, fit and transform the numerical columns, replace the original columns, and display the head of the dataframe to verify.



```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Identify numerical columns
numerical_cols = df_cleaned.select_dtypes(include=np.number).columns.tolist()

# Instantiate the StandardScaler
scaler = StandardScaler()

# Fit and transform the numerical columns
df_cleaned[numerical_cols] = scaler.fit_transform(df_cleaned[numerical_cols])

# Display the head of the DataFrame
display(df_cleaned.head())
```

## Encode categorical features (optional)

### Subtask:
Encode categorical features in the `df_cleaned` DataFrame.


**Reasoning**:
Import the necessary library and identify categorical columns.



```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = df_cleaned.select_dtypes(include='category').columns
print("Categorical columns to encode:")
print(categorical_cols)
```

**Reasoning**:
Instantiate and apply OneHotEncoder to the categorical columns and then concatenate the encoded features with the rest of the DataFrame.



```python
# Instantiate the encoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit and transform the categorical columns
encoded_features = encoder.fit_transform(df_cleaned[categorical_cols])

# Create a DataFrame from the encoded features with informative column names
encoded_df = pd.DataFrame(encoded_features, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate the original DataFrame (excluding original categorical columns) with the encoded DataFrame
df_cleaned = pd.concat([df_cleaned.drop(columns=categorical_cols), encoded_df], axis=1)

# Display the head of the modified DataFrame
display(df_cleaned.head())
```
